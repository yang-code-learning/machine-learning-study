```python
# apply a non-linear function to "make things complex"
# usually rectifying nonlinearity(relu) and rangens hyperbolicus(tanh)
line  = np.linspace(-3,3,100)
plt.plot(line, np.tanh(line), label="tanh")
plt.plot(line, np.maximum(line, 0), label="relu")
plt.legend(loc='best')
plt.xlabel("x")
plt.ylabel("relu(x), tanh(x)")
```
